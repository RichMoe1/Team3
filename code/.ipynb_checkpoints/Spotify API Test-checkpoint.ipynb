```python
# MUSIXMATCH
# source: https://developer.musixmatch.com/documentation/api-reference/track-lyrics-get
# python: https://github.com/hudsonbrendon/python-musixmatch
# https://github.com/guoguo12/billboard-charts

music = 'hot-100'
chart = billboard.ChartData(music,date="2017-12-31")

chart_list = []
music = 'hot-100'
chart = billboard.ChartData(music,date="2017-12-31")
for row in chart:
    chart_list.append(str(row).replace("'","",1))
chart_df = pd.DataFrame(chart_list)
chart_df = chart_df[0].str.split("' by ",expand=True)
chart_df = chart_df.rename(columns={0:"Song",1:"Artist"})
chart_df
# isinstance(chart_df[0][0],str)
# type(chart)
```
